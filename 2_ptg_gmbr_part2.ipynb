```python
import cv2
gambar = cv2.imread('/content/drive/MyDrive/Tugas_Akhir/new/IMG_20220613_082723.jpg')
```

```python
gambar.shape
```

```python
from PIL import Image

for i in range(0, 2250, 250):
  for j in range(0, 4000, 250):
    im = Image.fromarray(gambar[i:i+250,j:j+250])
    im.save(str(i)+'_'+str(j)+".jpg")
```
